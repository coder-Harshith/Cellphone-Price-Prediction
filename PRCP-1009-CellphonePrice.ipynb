# Task 2: Price Range Prediction
## Data preprocessing

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop('price_range', axis=1)
y = df['price_range']
```

### Split the dataset

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# Model training and evaluation

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
```

## Logistic regression

```python
## Hyperparameter tuning
# Logistic Regression
lr_model = LogisticRegression(max_iter=1000)
param_grid = {'C': [0.1, 1, 10]}
lr_grid = GridSearchCV(lr_model, param_grid, cv=5)
lr_grid.fit(X_train_scaled, y_train)
lr_preds = lr_grid.best_estimator_.predict(X_test_scaled)
print("Logistic Regression Report:\n", classification_report(y_test, lr_preds))
print('Accuracy Score:', accuracy_score(y_test, lr_preds))
# do it fromm here----------------------------------------------------------------------------
```

# Decision Tree

```python
# Decision Tree
dt_model = DecisionTreeClassifier()
dt_params = {'max_depth': [5, 10, 15], 'criterion': ['gini', 'entropy']}
dt_grid = GridSearchCV(dt_model, dt_params, cv=5)
dt_grid.fit(X_train, y_train)
dt_preds = dt_grid.best_estimator_.predict(X_test)
print("Decision Tree Report:\n", classification_report(y_test, dt_preds))
print('Accuracy Score:', accuracy_score(y_test, dt_preds))
```

# Random Forest

```python
# Random Forest
rf_model = RandomForestClassifier(random_state=42)
rf_params = {'n_estimators': [50, 100, 200], 'max_depth': [5, 10, None]}
rf_grid = GridSearchCV(rf_model, rf_params, cv=5)
rf_grid.fit(X_train, y_train)
rf_preds = rf_grid.best_estimator_.predict(X_test)
print("Random Forest Report:\n", classification_report(y_test, rf_preds))
print('Accuracy Score:', accuracy_score(y_test, rf_preds))
```

# XGboost

```python
# XGBoost
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
xgb_params = {'n_estimators': [50, 100, 200], 'max_depth': [3, 6, 9]}
xgb_grid = GridSearchCV(xgb_model, xgb_params, cv=5)
xgb_grid.fit(X_train, y_train)
xgb_preds = xgb_grid.best_estimator_.predict(X_test)
print("XGBoost Report:\n", classification_report(y_test, xgb_preds))
print('Accuracy Score:', accuracy_score(y_test, xgb_preds))
```

### Logistic Regression performed reasonably well but struggled with complex relationships.
### Decision Tree captured patterns better but was prone to overfitting.
### Random Forest provided the best balance between accuracy and generalization.
### XGBoost performed well but required hyperparameter tuning for further improvement.
